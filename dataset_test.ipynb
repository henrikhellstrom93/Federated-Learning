```python
#Imports
import models
import numpy as np
from matplotlib import pyplot
```

```python
#Simulation Variables
K = 2
```

```python
local_datasets, global_dataset = models.load_dataset(K)
labels = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]
```

```python
#Display 1 random image from the global training dataset
N = global_dataset[0].shape[0]
i = np.random.randint(N)
label_arg = np.argmax(global_dataset[1][i])
pyplot.imshow(global_dataset[0][i])
pyplot.xlabel(labels[label_arg])
```
