```python
#Imports
import models
import numpy as np
import time
import tensorflow as tf
from matplotlib import pyplot
```

```python
#Simulation Variables
K = 2
```

```python
labels = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]
model_list, global_model = models.setup_models(K, len(labels))
```

```python
#Global update (FedAvg uplink step)
l = 0 #Layer to print first weight from
print("One particular weight of global model =", global_model.layers[l].weights[0][0][0][0][0])
print("Same particular weight of local model 0 =", model_list[0].layers[l].weights[0][0][0][0][0])
print("Same particular weight of local model 1 =", model_list[1].layers[l].weights[0][0][0][0][0])
print("Perfoming global update...")
model_list, global_model = models.global_update(model_list, global_model)
print("Same particular weight of global model =", global_model.layers[l].weights[0][0][0][0][0])
global_weight = float(global_model.layers[l].weights[0][0][0][0][0])
local_weight0 = float(model_list[0].layers[l].weights[0][0][0][0][0])
local_weight1 = float(model_list[1].layers[l].weights[0][0][0][0][0])
if global_weight - (local_weight0/2+local_weight1/2) < 0.01:
    print("Global weight is arithmetic mean of local weights, looks good!")
else:
    print("Global weight is not what was expected! Check models.py")
```
