```python
#Imports
import models
import numpy as np
import time
import tensorflow as tf
from matplotlib import pyplot
```

```python
#Simulation Variables
K = 2
```

```python
#Download global dataset and split it into K local datasets
dataset = models.load_dataset(K)
#CIFAR10 labels
labels = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]
#Create K local models and one global model with randomly initialized weights
model_list, global_model = models.setup_models(K, len(labels))
```

```python
#Broadcast to devices
l = 0 #Layer to print first weight from
print("One particular weight of global model =", global_model.layers[l].weights[0][0][0][0][0])
print("Same particular weight of local model =", model_list[0].layers[l].weights[0][0][0][0][0])
print("Broadcasting global model...")
model_list, global_model = models.model_broadcast(model_list, global_model)
print("Same particular weight of local model =", model_list[0].layers[l].weights[0][0][0][0][0])
global_weight = float(global_model.layers[l].weights[0][0][0][0][0])
local_weight = float(model_list[0].layers[l].weights[0][0][0][0][0])
if global_weight == local_weight:
    print("Local and global weights are the same, seems good!")
else:
    print("Local and global weights are unequal! Check models.py file")
```
